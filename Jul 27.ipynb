```python
# Q: probs: [0.1, 0.2, 0.7] -> return 0 with 0.1, 1 with 0.2 and 2 with 0.7

def choose_action():
    r = random.random()
    if r < 0.1:
        return 0
    elif r < 0.2+0.1:
        return 1
    else:
        return 2
```

```python
def test_three():
    lc = 100000
    c1, c2, c3 = 0, 0, 0
    for i in range(lc):
        res = choose_action()
        if res == 0:
            c1 += 1
        elif res == 1:
            c2 += 1
        else:
            c3 += 1
    print(f"{c1/lc}, {c2/lc}, {c3/lc}, ")
```

```python
test_three()
```

```python
# Q: probs: [0.1, 0.2, 0.7] -> return 0 with 0.1, 1 with 0.2 and 2 with 0.7

def choose_action(probs):
    r = random.random()
    _probs = probs.copy()
    l = len(_probs)
    for i in range(1, l):
        _probs[i] += _probs[i-1]
    for i in range(l-1):
        if r < _probs[i]:
            return i
    return len(l) - 1
```

```python
probs = [0.1, 0.2 , .7] -> [0.1, 0.3, 1]
[0.7, 0.2 , .1] -> [0.7, 0.9, 1]
[0.7, 0.1 , .2] -> [0.7, 0.8, 1]
```
