### ‚úÖ **Step 1: Choose Your Base (Between 3 and 13)**

For example:

* If you choose base 5, your digits will be: 0, 1, 2, 3, 4

### ‚úÖ **Step 3: Add and Subtract Single-Digit Numbers in Your Base**

üëâ **Exercise 3A: Add all pairs of single-digit numbers in your base**

**Explanation:**
Add two digits and express the result in your base. If the sum is ‚â• base, carry over.

‚úÖ **Example in base 5**:

```
3 + 4 = 7 ‚Üí In base 5: 7 = 1 * 5 + 2 ‚Üí Answer = 12 (1 carry, 2 below)
```

üëâ **Exercise 3B: Subtract all possible pairs of digits**

**Explanation:**
Make sure the result is never negative.
If the first digit is smaller, skip or indicate "not allowed without borrowing".

‚úÖ **Example in base 6**:

```
5 - 2 = 3 ‚Üí In base 6: still 3
2 - 5 ‚Üí Not allowed without borrowing
```

‚úçÔ∏è **Your Task:**
Create two tables:

* Addition table: all combinations of digits
* Subtraction table: only where first digit ‚â• second digit



````python
# Prepare digits
digits = [str(i) if i < 10 else chr(ord('A') + i - 10) for i in range(base)]

# Addition Table
print("Addition Table (Base {})".format(base))
print("  + | " + " | ".join(digits) + " |")
print("-" * (5 + 4 * base))
for d1 in range(base):
    row = [digits[d1]]
    for d2 in range(base):
        s = d1 + d2
        if s < base:
            row.append(digits[s])
        else:
            # Convert to base b
            carry = s // base
            rem = s % base
            val = digits[carry] + digits[rem]
            row.append(val)
    print("  {} | ".format(digits[d1]) + " | ".join(row[1:]) + " |")

print("\nSubtraction Table (Base {})".format(base))
print("  - | " + " | ".join(digits) + " |")
print("-" * (5 + 4 * base))
for d1 in range(base):
    row = [digits[d1]]
    for d2 in range(base):
        if d1 >= d2:
            row.append(digits[d1 - d2])
        else:
            row.append("‚Äì")
    print("  {} | ".format(digits[d1]) + " | ".join(row[1:]) + " |")
````

### ‚úÖ **Step 4: Add and Subtract Multi-Digit Numbers**

üëâ **Exercise 4A: Add two 2-digit numbers in your base**

**Example in base 6**:

```
35‚ÇÜ + 24‚ÇÜ
Step-by-step:
5 + 4 = 9 ‚Üí 9 = 1 * 6 + 3 ‚Üí Write 3, carry 1
3 + 2 + 1 (carry) = 6 ‚Üí 6 = 1 * 6 + 0 ‚Üí Write 0, carry 1
Result: 103‚ÇÜ
```

üëâ **Exercise 4B: Subtract two 2-digit numbers**

**Example in base 7**:

```
52‚Çá - 34‚Çá ‚Üí Convert digits, subtract with borrowing if needed
```

‚úçÔ∏è **Your Task:**
Solve at least 3 addition and 3 subtraction problems involving two-digit numbers.

````python
# Helper to convert digit character to value
def char_to_value(char):
    if char.isdigit():
        return int(char)
    else:
        return ord(char) - ord('A') + 10

# Helper to convert value to digit character
def value_to_char(value):
    if value < 10:
        return str(value)
    else:
        return chr(ord('A') + value - 10)

# Addition problems
addition_problems = [
    ['1', '5', '2', '4'],  # 15 + 24
    ['3', '7', '1', '9'],  # 37 + 19 (if base allows)
    ['2', '8', '3', '6'],  # 28 + 36 (if base allows)
]

addition_count = 0
for problem in addition_problems:
    if addition_count >= 3:
        break
    
    # Check if all digits are valid for this base
    valid = True
    for digit_char in problem:
        digit_val = char_to_value(digit_char)
        if digit_val >= base:
            valid = False
            break
    
    if not valid:
        continue
    
    addition_count += 1
    a1, a0 = problem[0], problem[1]  # First number
    b1, b0 = problem[2], problem[3]  # Second number
    
    print(f"\nProblem {addition_count}: {a1}{a0} + {b1}{b0} (base {base})")
    print("-" * 40)
    
    # Convert to values
    a1_val = char_to_value(a1)
    a0_val = char_to_value(a0)
    b1_val = char_to_value(b1)
    b0_val = char_to_value(b0)
    
    # Step 1: Add ones place
    ones_sum = a0_val + b0_val
    print(f"Step 1: {a0} + {b0} = {ones_sum}")
    
    if ones_sum >= base:
        ones_digit = ones_sum % base
        carry1 = ones_sum // base
        print(f"  ‚Üí {ones_sum} = {carry1} * {base} + {ones_digit}")
        print(f"  ‚Üí Write {value_to_char(ones_digit)}, carry {carry1}")
    else:
        ones_digit = ones_sum
        carry1 = 0
        print(f"  ‚Üí Write {value_to_char(ones_digit)}, no carry")
    
    # Step 2: Add tens place
    tens_sum = a1_val + b1_val + carry1
    print(f"Step 2: {a1} + {b1} + {carry1} (carry) = {tens_sum}")
    
    if tens_sum >= base:
        tens_digit = tens_sum % base
        carry2 = tens_sum // base
        print(f"  ‚Üí {tens_sum} = {carry2} * {base} + {tens_digit}")
        print(f"  ‚Üí Write {value_to_char(tens_digit)}, carry {carry2}")
        result = f"{value_to_char(carry2)}{value_to_char(tens_digit)}{value_to_char(ones_digit)}"
    else:
        tens_digit = tens_sum
        carry2 = 0
        print(f"  ‚Üí Write {value_to_char(tens_digit)}, no carry")
        result = f"{value_to_char(tens_digit)}{value_to_char(ones_digit)}"
    
    print(f"Result: {a1}{a0} + {b1}{b0} = {result} (base {base})")

print(f"\n{'='*60}")


# Subtraction problems
subtraction_problems = [
    ['5', '2', '3', '4'],  # 52 - 34
    ['4', '1', '2', '5'],  # 41 - 25
    ['6', '0', '1', '7'],  # 60 - 17 (if base allows)
]

subtraction_count = 0
for problem in subtraction_problems:
    if subtraction_count >= 3:
        break
    
    # Check if all digits are valid for this base
    valid = True
    for digit_char in problem:
        digit_val = char_to_value(digit_char)
        if digit_val >= base:
            valid = False
            break
    
    if not valid:
        continue
    
    a1, a0 = problem[0], problem[1]  # First number (minuend)
    b1, b0 = problem[2], problem[3]  # Second number (subtrahend)
    
    # Convert to decimal to check if subtraction is valid
    a_decimal = char_to_value(a1) * base + char_to_value(a0)
    b_decimal = char_to_value(b1) * base + char_to_value(b0)
    
    if a_decimal < b_decimal:
        continue  # Skip if result would be negative
    
    subtraction_count += 1
    print(f"\nProblem {subtraction_count}: {a1}{a0} - {b1}{b0} (base {base})")
    print("-" * 40)
    
    # Convert to values
    a1_val = char_to_value(a1)
    a0_val = char_to_value(a0)
    b1_val = char_to_value(b1)
    b0_val = char_to_value(b0)
    
    # Step 1: Subtract ones place
    print(f"Step 1: {a0} - {b0}")
    
    if a0_val >= b0_val:
        ones_diff = a0_val - b0_val
        borrow = 0
        print(f"  ‚Üí {a0_val} - {b0_val} = {ones_diff}")
        print(f"  ‚Üí Write {value_to_char(ones_diff)}, no borrowing")
    else:
        ones_diff = (a0_val + base) - b0_val
        borrow = 1
        print(f"  ‚Üí Need to borrow: {a0_val} + {base} - {b0_val} = {ones_diff}")
        print(f"  ‚Üí Write {value_to_char(ones_diff)}, borrow 1")
    
    # Step 2: Subtract tens place
    print(f"Step 2: {a1} - {b1} - {borrow} (borrow)")
    
    tens_diff = a1_val - b1_val - borrow
    print(f"  ‚Üí {a1_val} - {b1_val} - {borrow} = {tens_diff}")
    print(f"  ‚Üí Write {value_to_char(tens_diff)}")
    
    # Remove leading zero if present
    if tens_diff == 0:
        result = value_to_char(ones_diff)
    else:
        result = f"{value_to_char(tens_diff)}{value_to_char(ones_diff)}"
    
    print(f"Result: {a1}{a0} - {b1}{b0} = {result} (base {base})")

print(f"\n{'='*60}")
````

### ‚úÖ **Step 5: Write Multiplication Table in Your Base**

üëâ **Exercise 5:**
Create the **multiplication table** for your base up to (b - 1) √ó (b - 1)

**Explanation:**
Multiply each pair of digits, then convert the result to your base.

‚úÖ **Example in base 4**:

```
2 √ó 3 = 6 ‚Üí In base 4: 6 = 1 * 4 + 2 ‚Üí Answer = 12
```

‚úçÔ∏è **Your Task:**
Make a full multiplication table (rows and columns from 0 to b-1)

---

````python
# Convert decimal number to base representation
def to_base_representation(num, base_val):
    if num == 0:
        return '0'
    
    result = ''
    while num > 0:
        digit = num % base_val
        result = value_to_char(digit) + result
        num //= base_val
    
    return result

# Create header row
print(f"MULTIPLICATION TABLE FOR BASE {base}")
print(f"{'='*80}")
header = "   √ó"
for j in range(base):
    header += f" {digits[j]:>4}"
print(header)
print("   " + "-" * (5 * base + 1))

# Generate multiplication table
multiplication_data = []
for i in range(base):
    row_str = f" {digits[i]} |"
    row_data = []
    
    for j in range(base):
        # Calculate product in decimal
        product_decimal = i * j
        
        # Convert to base representation
        product_base = to_base_representation(product_decimal, base)
        
        # Format for display
        row_str += f" {product_base:>4}"
        
        # Store data for analysis
        row_data.append({
            'digit1': digits[i],
            'digit2': digits[j],
            'decimal_product': product_decimal,
            'base_product': product_base,
            'is_multi_digit': len(product_base) > 1
        })
    
    print(row_str)
    multiplication_data.append(row_data)
````



### ‚úÖ **Step 6: Multiply Two-Digit Numbers in Your Base**

üëâ **Exercise 6:**
Multiply two 2-digit numbers in your base

**Example in base 5**:
Multiply 13‚ÇÖ √ó 21‚ÇÖ
Steps:

1. Multiply digits as in long multiplication
2. Convert intermediate products to base 5
3. Add all rows (in base 5)

‚úÖ **Step-by-step (summary)**:

* Multiply ones place
* Multiply tens place and shift one place left
* Add rows in base

‚úçÔ∏è **Your Task:**
Multiply at least 2 pairs of 2-digit numbers in your base.

````python
# Get two 2-digit numbers from user
print(f"Enter two 2-digit numbers in base {base}")
print(f"Valid digits: {', '.join(digits)}")

problems = []
for i in range(2):
    while True:
        num = input(f"Enter 2-digit number {i+1}: ").strip().upper()
        if len(num) == 2:
            # Check if all digits are valid for this base
            valid = all(char_to_value(d) < base for d in num)
            if valid:
                problems.append([num[0], num[1]])
                break
            else:
                print(f"Invalid digits for base {base}. Use only: {', '.join(digits)}")
        else:
            print("Please enter exactly 2 digits")

for prob_num, problem in enumerate(problems, 1):
    a1, a0 = problem  # First number
    if prob_num == 1:
        b1, b0 = problems[1]  # Second number
    else:
        continue  # Only do first multiplication
    
    # For the multiplication, use both numbers
    a1, a0 = problems[0]  # First number
    b1, b0 = problems[1]  # Second number
    
    print(f"\nProblem: {a1}{a0} √ó {b1}{b0} (base {base})")
    print("=" * 40)
    
    # Convert to values
    a1_val, a0_val = char_to_value(a1), char_to_value(a0)
    b1_val, b0_val = char_to_value(b1), char_to_value(b0)
    
    # Step 1: Multiply by ones place (b0)
    print(f"Step 1: Multiply {a1}{a0} √ó {b0}")
    
    # a0 √ó b0
    prod1 = a0_val * b0_val
    prod1_base = to_base_representation(prod1, base)
    print(f"  {a0} √ó {b0} = {prod1} = {prod1_base}‚Çç{base}‚Çé")
    
    # a1 √ó b0 (shifted left)
    prod2 = a1_val * b0_val
    prod2_base = to_base_representation(prod2, base)
    print(f"  {a1} √ó {b0} = {prod2} = {prod2_base}‚Çç{base}‚Çé (shift left ‚Üí {prod2_base}0)")
    
    # First partial product
    first_partial = prod1 + prod2 * base
    first_partial_base = to_base_representation(first_partial, base)
    print(f"  First row: {first_partial_base}‚Çç{base}‚Çé")
    
    # Step 2: Multiply by tens place (b1) and shift
    print(f"\nStep 2: Multiply {a1}{a0} √ó {b1} (shift left)")
    
    # a0 √ó b1 (shifted left)
    prod3 = a0_val * b1_val
    prod3_base = to_base_representation(prod3, base)
    print(f"  {a0} √ó {b1} = {prod3} = {prod3_base}‚Çç{base}‚Çé (shift left ‚Üí {prod3_base}0)")
    
    # a1 √ó b1 (shifted left twice)
    prod4 = a1_val * b1_val
    prod4_base = to_base_representation(prod4, base)
    print(f"  {a1} √ó {b1} = {prod4} = {prod4_base}‚Çç{base}‚Çé (shift left twice ‚Üí {prod4_base}00)")
    
    # Second partial product
    second_partial = (prod3 + prod4 * base) * base
    second_partial_base = to_base_representation(second_partial, base)
    print(f"  Second row: {second_partial_base}‚Çç{base}‚Çé")
    
    # Step 3: Add partial products
    print(f"\nStep 3: Add partial products in base {base}")
    print(f"    {first_partial_base}")
    print(f"  + {second_partial_base}")
    print("  " + "-" * max(len(first_partial_base), len(second_partial_base)))
    
    final_result = first_partial + second_partial
    final_result_base = to_base_representation(final_result, base)
    print(f"    {final_result_base}")
    
    print(f"\nResult: {a1}{a0} √ó {b1}{b0} = {final_result_base} (base {base})")
````
