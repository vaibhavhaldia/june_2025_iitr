```python
# Using recursion
def multiply_recur(a, b):
    print(f"multiply_recur({a}, {b})")
    if b == 0:
        return 0
    return a + multiply_recur(a, b-1)
```

```python
multiply_recur(0, 4)
```

```python
# Using recursion
def multiply_recur(a, b):
    print(f"multiply_recur({a}, {b})")
    if b == 0:
        return 0
    print("Calling itself and will wait for computation")
    r = multiply_recur(a, b-1)
    print(f"Finished calulating r: {r}")
    return a + r
```

```python
multiply_recur(4, 3)
```

```python
multiply_recur(4, 3000)
```

```python
# Lambda Calculus
```

```python
def multiply_recur0(a):
    return 0
def multiply_recur1(a):
    return a + multiply_recur0(a)

def multiply_recur2(a):
    return  a + multiply_recur1(a)
def multiply_recur3(a):
    return  a + multiply_recur2(a)
def multiply_recur4(a):
    return a + multiply_recur3(a)
```

```python
multiply_recur4(4)
```

```python
# Using recursion
def multiply_recur(a, b, ind=""):
    print(f"{ind} multiply_recur({a}, {b})")
    if b == 0:
        print(f"{ind} Returning {0}")
        return 0
    r = multiply_recur(a, b-1, ind + "\t")
    ans = a + r
    print(f"{ind} Returning {ans}")
    return ans
```

```python
multiply_recur(4, 3)
```

```python
def pow_loop(a, b): # a to the power b
    result = 1
    for i in range(b):
        result = result * a
    return result
```

```python
def pow_rec(a, b, ind=''):
    print(f"{ind}pow_rec({a}, {b})")
    result = 1
    if b == 0:
        print(f"{ind}result: {result}")
        return result
    result = a * pow_rec(a, b - 1, ind + '\t')
    print(f"{ind}result: {result}")
    return result
```

```python
# Tower of Hanoi
# https://www.mathsisfun.com/games/towerofhanoi.html
```

```python
def move1(n, source, target):
    print(f"Actually moving {n} from {source} to {target}")
```

```python
def move2(n, source, aux, target):
    move1(n-1, source, aux)
    move1(n, source, target)
    move1(n-1, aux, target)
```

```python
move2(2, "SRC", "AUXX", "TGT")
```

```python
def print_move(n, source, target, ind):
    print(f"{ind}Actually Moving {n} from {source} to {target}")
```

```python
# For three discs
def move(n, src, aux, target, ind=''):
    print(f"{ind}move({n}, {src}, {aux}, {target})")
    if n == 1:
        print_move(n, src, target, ind)
    else:
        move(n-1, src, target, aux, ind+'  ')
        print_move(n, src, target, ind)
        move(n-1, aux, src, target, ind+'  ')
```

```python
move(1, 'A', 'B', 'C')
```
